# reading the given csv dataframe 

```python
import pandas as pd 
import numpy as np 
data=pd.read_csv("V:/assign4.csv")
data
```

# #removing the null rows in the dataframe 

```python
data.dropna()
```

# filling the empty  cells of column charges with mean median mode 

```python
x=data.charge.mean()
x
```

```python
y=data.charge.median()
y
```

```python
z=data.charge.mode()
z
```

```python
data.isnull().sum()
```

```python
data.charge.fillna(x,inplace=True)
```

```python
data.charge.fillna(y,inplace=True)
```

```python
data.charge.fillna(z,inplace=True)
```

```python
data.replace(to_replace=data.ages>60,value=50)
```
